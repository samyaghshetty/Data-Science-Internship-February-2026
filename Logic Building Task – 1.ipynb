1.User Login Check

```python
username = "admin"
password = "1234"

username_input = input("Enter username: ")
password_input = input("Enter password: ")

if username_input == username and password_input == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2.Pass/Fail Analyzer

```python
marks = [45,78,90,33,60]
passed = 0
failed = 0
for i in  marks:
    if i>=50:
        passed+=1
    else:
        failed+=1
print(f"Total number of students: {len(marks)}\nNumber of students passed are : {passed}\nNumber of students failed are : {failed}")
```

3.Simple Data Cleaner

```python
names = ["Alice","bob","CHARLIE"]
cleaned_names = []
for i in names:
    i = i.strip() #This removes extra space
    i = i.lower() #Convert to lowercase
    cleaned_names.append(i)
print(cleaned_names)
```

4.Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for i  in messages:
    print("Message :",i)
    print("Length :",len(i))
    if len(i)>10:
        print("Status : Long Message")
    else:
        print("Status : Short Message")
    print()
```

5.Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
    if i=="ERROR":
        count+=1
print("Total ERROR count:",count)
```
