## 1. Employee Performance Bonus Eligibility

```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

highest = 0
# Find the highest score among all employees
for score in employees.values():
    if score > highest:
        highest = score
        
top_performer = []
# Find employees who have the highest score
for name,score in employees.items():
    if score==highest:
        top_performer.append(name)
        
print("Top Performers Eligible for Bonus:",",".join(top_performer),
      f"(Score: {highest})")
```

## 2. Search Query Keyword Analysis

```python
website = "Buy mobile phone buy phone online".lower()

for ch in ",.!?":
    website = website.replace(ch,"")
    
words = website.split()
freq = {}

# Count frequency of each word
for word in words:
    if word in freq:
        freq[word]+=1
    else:
        freq[word]=1
        
result = {}

# Store words that appear more than once
for word,count in freq.items():
    if count>1:
        result[word]=count
print(result)
```

## 3. Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
# Empty list to store valid (even) readings
even_readings = [] 

for hours,readings in enumerate(sensor_readings):
    if readings%2==0: # Check if the sensor reading is even
        even_readings.append((hours,readings))
        
print(f"Valid Sensor Readings (Hour, Value):{even_readings}")
```

## 4. Email Domain Usage Analysis

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
gmail = 0
yahoo = 0
total = 0
# Loop through each email address
for mail in emails:
    total+=1
    # Check if the email ends with gmail.com
    if mail.endswith("gmail.com"):
        gmail+=1
    # Otherwise, count it as a yahoo email
    else:
        yahoo+=1
print(f"gmail.com: {gmail/total*100}%\nyahoo.com: {yahoo/total*100}%")
```

## 5. Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
daily_sales = []
total = 0
# Calculate total sales
for i in sales:
    total+=i
# Calculate average sales
average = total/len(sales)
# Find days where sales are more than 30% above average
for day,value in enumerate(sales, start=1):
    if average*1.3<value:
        daily_sales.append((day,value))
print(f"Day {day} -> {value}")
```

## 6. Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
count = {}
# Count frequency
for id in user_ids:
    if id in count:
        count[id]+=1
    else:
        count[id]=1
# Store user that appear more than once
for user,times in count.items():
    if times>1:
        print(f"{user} -> {times} times")
```
