## 1.Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount >= 50 and amount in valid_plans:
        return True
    return False

while True:
    amt = int(input("Enter recharge amount: ₹"))
    if validate_recharge(amt):
        print(f"Recharge of ₹{amt} accepted")
        break
    else:
        print(f"₹{amt} is an invalid amount. Try again!")
```

## 2.Inventory Reorder Alert System

```python
def check_stock(products):
    for product,stock in products.items():
        if stock < 15:
            print(f"{product}:Reorder Alert (Stock = {stock})")
        else:
            print(f"{product}: Stock Ok (Stock = {stock})")
inventory = {
    "Apple":10,
    "Mango":20
}
check_stock(inventory)
```

## 3.Student Result Processing System

```python
def student_result(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total/len(marks)
    if average >=50:
        return "Pass"
    else:
        return "Fail"
student_marks = [60, 55, 70, 45, 50]
result = student_result(student_marks)
print("Result:", result)
```

## 4.Cab Fare Estimator with Retry Option

```python
def cab(km, peak_hour):
    base_fare = 50
    total = base_fare + (12*km)
    if peak_hour == 'yes':
        total += total * 0.25
    return total
distance = float(input("Enter distance in km: "))
peak = input("Is it peak hour? (yes/no): ").lower()

fare = cab(distance, peak)
print("Total Cab Fare: ₹", fare)
```

## 5.Employee Attendance Eligibility

```python
def check_attendance(attendance_list):
    present_count = 0
    
    for day in attendance_list:
        if day == "P":
            present_count += 1

    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

result = check_attendance(attendance)
print("Attendance Status:", result)
```

## 6.Password Strength Checker

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
password = input("Enter password: ")
result = check_password_strength(password)
print(result)
```
