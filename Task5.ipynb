## 1. Social Media – Post Engagement Analyzer

```python
def analyze_post(likes_list):
    total_likes = 0

    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


likes_data = [200, 350, 150, 400] 
analyze_post(likes_data)
```

## 3. Agriculture – Rainfall Adequacy Checker

```python
def adequacy(rainfall,required_level):
    total_rain = 0
    for rain in rainfall:
        total_rain += rain
    average = total_rain/len(rainfall)
    if average>=required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)
seasonal_rainfall = [70, 75, 68, 80, 67]
required_level = 70
adequacy(seasonal_rainfall, required_level)
    
```

## 4. Social Media – Duplicate Account Detection

```python
def duplicate_account(username):
    unique_users = set(username)
    if len(username) != len(unique_users):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
usernames_list = ["sam", "alex", "john", "sam", "maria"]
duplicate_account(usernames_list)
```

## 6. Agriculture – Premium Crop Price Filter

```python
def price_filter(prices):
    premium_crops = []
    for price in prices:
        if price>2000:
            premium_crops.append(price)
    print("Premium Crops:",premium_crops)
crop_price=[1000,1500,2000,2500,3200]
price_filter(crop_price)
```

## 7. System Monitoring – Application Health

```python
def health_monitoring(errors):
    if errors == 0:
        print("Error Count:",errors)
        print("System Status: Healthy")
    elif errors <= 5:
        print("Error Count:",errors)
        print("System Status: Minor Issues")
    else :
        print("Error Count:",errors)
        print("System Status: Critical Issues")
health_monitoring(7)
```

## 10. Smart Electricity Bill Analyzer

```python
def bill_analyzer(units):
    total_bill = 0
    if units<=100:
        total_bill += units*3
        print("Total Bill:",total_bill)
    elif 101<=units<=200:
        total_bill += (100 *3)+(units-100)*5
        print("Total Bill:",total_bill)
    else:
        total_bill += (100*3)+(100*5)+(units-200)*7
        print("Total Bill:",total_bill)
    if total_bill <500 :
        print("Low Usage")
    elif 500<=total_bill<=1500:
        print("Moderate Usage")
    else:
        print("High Usage")
bill_analyzer(170)
        
```
