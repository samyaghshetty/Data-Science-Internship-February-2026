1.Unique Words in a Sentence

```python
sentence = "python is easy and python is powerful"

words = sentence.split()

unique_words = set(words)

# Display results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

2.Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_salary = 0
highest_name = ""
for name in employees:
    salary = employees[name]
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

3.Find the maximum and minimum values from a list of numbers

```python
numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for i in range(1, len(numbers)):
    if numbers[i] > maximum:
        maximum = numbers[i]
    if numbers[i] < minimum:
        minimum = numbers[i]

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

4.Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
count = 0

for i in prices:
    if i > 1000:
        count += 1

print("Products above 1000:", count)
```

5.Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

```python
attendance = ["P", "P", "A", "P", "P"]

present = 0
total = 0
for status in attendance:
    if status == "P":
        present += 1
        total += 1
    else:
        total += 1

print("Attendance Percentage:", (present / total) * 100)
```

6.Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_dict = {}
for number in phone_numbers:
    if number not in unique_dict:
        unique_dict[number] = 1

print("Unique phone numbers:", list(unique_dict.keys()))
```

7.Count Character Frequency

```python
text = "ppython"

char_count = {}

for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```
